```python
import requests
import json

def get_businesses1(query,api_key,min_rating,max_results):
    a=[]
    b=()
    count=1
    api_key='Insert API Key'
    request_page=1
    
    while count<=max_results:
        if request_page==1:
            url="https://maps.googleapis.com/maps/api/place/textsearch/json?query=%s&key=%s" % (query,api_key)
            response = requests.get(url)
            if not response.status_code == 200:
                print("HTTP error",response.status_code)
            restaurants=response.json()
        else:
            try:
                restaurants['next_page_token']
                newPageToken = restaurants['next_page_token']
                new_url = "https://maps.googleapis.com/maps/api/place/textsearch/json?pagetoken=%s&query=%s&key=%s" % (newPageToken,query,api_key)
                response = requests.get(new_url)
                if not response.status_code == 200:
                    print("HTTP error",response.status_code)
                restaurants=response.json()

                while response.json().get('status') != 'OK':
                    from time import sleep
                    from random import random
                    sleep(random())
                    response = requests.get(new_url)
                    restaurants = response.json()
            except:
                return a
            
        for i in restaurants['results']:
            x=['opening_hours','price_level','rating']
            for element in x:
                try:
                    i[element]
                except:
                    if element=='rating':
                        i[element]=0
                    else:
                        i[element]=None

            if i['rating']>=min_rating:
                b=(i['name'],i['formatted_address'],i['opening_hours'],i['price_level'],i['rating'])
                a.append(b)
                count+=1
               
            
            if count>max_results:
                break

        request_page+=1
    
    return a   
```
